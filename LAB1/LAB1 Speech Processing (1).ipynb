## A1.Load the recorded speech file into your python workspace. Once loaded, plot the graph for the speech signal.You may use the below code from librosa asa reference.

```python
import librosa
y, sr = librosa.load("audio.wav")
librosa.display.waveshow(y)
```

## A2. Observe the length & magnitude range of the signalfrom theplot. Observing the plot, relate tothe spoken words and silence between the words.

```python
import numpy as np
```

```python
print("Length of the signal:",len(y))
print("Magnitude range:",np.min(y),"to", np.max(y))
```

## A3. Take a small segment of the signal and play it. 

```python
import sounddevice as sd
```

```python
start_time = 1
end_time = 1.25
start_sample = int(start_time * sr)
end_sample = int(end_time * sr)
segment = y[start_sample:end_sample]

librosa.display.waveshow(segment)

sd.play(segment, sr)
sd.wait()
```
